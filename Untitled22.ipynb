```python
# Training a Linear Regression Model
# X and y List
X = df[['Income', 'Age of house', 'No of rooms', 'no of bedrooms',
       'population ']]

y = df['Price']
```

```python
# Split Data into Train, Test

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
#Creating and Training the LinearRegression Model

from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
# LinearRegression Model Evaluation

print(lm.intercept_)
```

```python

from sklearn.preprocessing import StandardScaler
```

```python
# Standardize the predictor variables using the StandardScaler:
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
# Fit a linear regression model on the standardized predictor variables:
model = LinearRegression()
model.fit(X_std, y)
```

```python
# Get the standardized coefficients from the model:
coefficients = model.coef_
```

```python
# Calculate the feature importance by taking the absolute value of the standardized coefficients and normalizing them:
feature_importance = np.abs(coefficients)
feature_importance /= feature_importance.sum()
```

```python
import pandas as pd

# create a dataframe with the feature names and importance values
df_feature_importance = pd.DataFrame({'Feature': X.columns, 'Importance': feature_importance})

# sort the dataframe by importance values in descending order
df_feature_importance = df_feature_importance.sort_values('Importance', ascending=False)

# reset the index of the dataframe
df_feature_importance = df_feature_importance.reset_index(drop=True)

# print the dataframe
print(df_feature_importance)
```

```python
import matplotlib.pyplot as plt

# plot the feature importances as horizontal bar graph
plt.barh(range(X.shape[1]), feature_importance, align='center')
plt.yticks(range(X.shape[1]), X.columns)
plt.xlabel('Normalized Importance')
plt.ylabel('Predictor Variable')
plt.show()
```

```python
# standardize the test set using the same scaler object used for training data
X_test_std = scaler.transform(X_test)

# predict house prices for the test set using the linear regression model
y_pred = model.predict(X_test_std)

# create a pandas dataframe with the predicted values and actual values for comparison
df_results = pd.DataFrame({'Actual Price': y_test, 'Predicted Price': y_pred})

# print the dataframe
print(df_results)
```
